### Data Assessment 

#### Visual Assessment
The 3 dataframe were first assessed visually using the df.sample method. Few obvious data quality issues were identified, and these include
- Missing values in most columns of the twitter_archive dataframe
- Unusual characters in the name column of twitter_archive dataframe
- Number of rows in all 3 dataframe did not match

#### Programmatic Assessment
The following functions were used to quickly assess all 3 dataframe
- Pd.head()
- Pd.info()
- Pd.describe()
- Pd.isnull().sum()
- Pd.[tweet_id).duplicated().sum()

Visual and programmatic assessment yielded the following data quality and tidiness issues.

##### Quality Issues 
- twitter_archive['name'] has some unusual lower-cased names
- twitter_archive['Timestamp'] data type is listed as an object instead of datetime, split into 2 columns(date, time)
- Source is a hyperlink
- tweet_id for all 3 tables is listed as int datatype
- Analysis requires actual tweets and not retweets. Rows that are retweets will be dropped   
retweeted_status_id, retweeted_status_timestamp and retweeted_status_user_id are not needed for the analysis and therefore will be dropped.
- in_reply_to_status_id and in_reply_to_user_id are listed as floats
- p1, p2, p3 columns in image_prediction contains unusual character(_)
- Inconsistent casing in p1, p2, p3
- total rating column missing

##### Tidiness Issues
- doggo, floofer, puppo and pupper columns need not be in separate columns
- twitter_archive_1 contains columns that required for analysis and therefore will be dropped.
