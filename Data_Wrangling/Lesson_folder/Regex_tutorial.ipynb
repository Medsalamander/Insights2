.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)

```python
import re
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
coreyms.com
321-555-4321
123.555.1234
123*555*1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T

cat
mat
pat
bat

'''
emails = '''
CoreyMSchafer@gmail.com
corey.schafer@university.edu
corey-321-schafer@my-work.net
'''

sentence = 'Start a sentence and then bring it to an end'
# to se
pattern = re.compile(r'abc')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# the search returned a match from indexes 1-4. to confirm this using string slicing
print(text_to_search[1:4])
```

- special characters . ^ $ * + ? { } [ ] \ | ( ) must be escaped with a backslash when being used

```python
# for example. to search for a url coreyms.com
pattern = re.compile(r'coreyms\.com')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#. returns any xter except new line
pattern = re.compile(r'.')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \d returns any difit between 0 and 9
pattern = re.compile(r'\d')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \D returns non digit. does oppposite of small letter d
pattern = re.compile(r'\D')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \w      - Word Character (a-z, A-Z, 0-9, _)
pattern = re.compile(r'\w')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \W returns not a word xter. opposite of \w
pattern = re.compile(r'\W')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \s      - Whitespace (space, tab, newline)
pattern = re.compile(r'\s')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \S      -  NOT Whitespace (space, tab, newline)
pattern = re.compile(r'\S')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \b      - Word Boundary - SPace or non-alpha numeric
pattern = re.compile(r'\bHa')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \B      not  Word Boundary - SPace or non-alpha numeric
pattern = re.compile(r'\BHa')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# ^       - Beginning of a String
pattern = re.compile(r'^Start')
matches = pattern.finditer(sentence)

for match in matches:
    print(match)
```

```python
# $       - End of a String
pattern = re.compile(r'end$')
matches = pattern.finditer(sentence)

for match in matches:
    print(match)
```

```python
# to match phone numbers
pattern = re.compile(r'\d\d\d.\d\d\d.\d\d\d\d')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# to match phone numbers with specific seperators, we use xter set which uses square brackets enclosing desired seperators
pattern = re.compile(r'\d\d\d[-.]\d\d\d[-.]\d\d\d\d')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# to match 800 or 900 numbers
pattern = re.compile(r'[89]00[-.]\d\d\d[-.]\d\d\d\d')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# the - is a special xter when put at the beginning or end it will match the literal - xter. but when placed between values it can specify a range of values.
pattern = re.compile(r'[1-5]')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# also for letters
pattern = re.compile(r'[a-z]')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# selecting both upper and lower case letters
pattern = re.compile(r'[a-zA-Z]')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# ^ in front of a xter set negates everything in the xter set
pattern = re.compile(r'[^a-zA-Z]')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# to match cat mat pat but not bat
pattern = re.compile(r'[^b]at')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

### we use quantifiers to match more than one xter at once
Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)


```python
# back to the phone number pattern
pattern = re.compile(r'\d{3}.\d{3}.\d{4}')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# to match names that start with mr
pattern = re.compile(r'Mr\.')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# to also return mr that is without period
pattern = re.compile(r'Mr\.?')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# to match upper case letters of names following mr
pattern = re.compile(r'Mr\.?\s[A-Z]')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# TO return the full names
pattern = re.compile(r'Mr\.?\s[A-Z]\w+')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

this only returned only schafer and smith but not T. TO resolve this we use the * quantifier

```python
pattern = re.compile(r'Mr\.?\s[A-Z]\w*')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# groups allow us to match several different patterns. To create group we use parenthesis
# to match mr, mrs ms

pattern = re.compile(r'M(r|s|rs)\.?\s[A-Z]\w*')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# OR
pattern = re.compile(r'(Mr|Ms|Mrs)\.?\s[A-Z]\w*')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
urls = '''
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
'''

```

```python
pattern = re.compile(r'https?://(www\.)?\w+\.\w+')
matches = pattern.finditer(urls)

for match in matches:
        print(match)
```

```python
# to capture domain names
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)

for match in matches:
        print(match)
```

```python
# to print group 0 which is the entire thing
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)

for match in matches:
        print(match.group(0))
```

```python
# to print group 1 which is the optional www
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)

for match in matches:
        print(match.group(1))
```

```python
# to print group 2
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)

for match in matches:
        print(match.group(2))
```

```python
# to print group 3 which is the top level domain
pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')
matches = pattern.finditer(urls)

for match in matches:
        print(match.group(3))
```

```python
# back reference

subbed_urls =  pattern.sub(r'\2\3', urls)

print(subbed_urls)
```
