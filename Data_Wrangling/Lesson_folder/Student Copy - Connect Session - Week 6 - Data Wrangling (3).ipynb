## Understanding Loops in Python
![for loops](https://e6v4p8w2.rocketcdn.me/wp-content/uploads/2020/06/Python-For-Loop-Structure.png)

### Loops - 2 types, For Loops & While Loops

Loops are important in Python or in any other programming language as they help you to execute a block of code repeatedly. 

### For Loops
A for loop is a programming concept that, when it's implemented, executes a piece of code over and over again "for" a certain number of times, based on a sequence. Unlike to the while loop, there isn't any condition actively involved - you just execute a piece of code repeatedly for a number of times. 

**Examples**

````python
# Print "Thank you" 5 times

for number in range(5):
    print("Thank you "+ str(number))
````

````python
# print out languages in list


languages = ['R', 'Python',  'Scala', 'Java', 'Julia', 'Rebecca']

for number in range(len(languages)): # the list items are 5 tiems
   print('Current language:', languages[number])
````

### While Loops

A while loop is a programming concept that, when it's implemented, executes a piece of code over and over again while a given condition ***still holds true.***

**Examples**

````python
# Take user input
number = 1

# Condition of the while loop
while number < 5 :   # defines the range
    print("Thank you")
    # Increment the value of the variable "number by 1"
    number = number+1
````

````python
# Take user input
number = 2 

# Condition of the while loop
while number < 5 :  
    # Find the mod of 2
    if number%2 == 0:  
        print("The number "+str(number)+" is even")
    else:
        print("The number "+str(number)+" is odd")

    # Increment `number` by 1
    number = number+1
````

````python
# define a function to use for a while loop

def collatz(number):
    # Is the mod of 2 equal to 0?
    if number % 2 == 0:
        print(number // 2)
        return number // 2

    # If the mod of 2 isn't equal to 0, print `3 * number + 1`
    elif number % 2 == 1:
        result = 3 * number + 1
        print(result)
        return result

# Ask input from the user    
n = input("Give me a number: ")

# As long as `n` is not equal to `1`, run `collatz()`
while n != 1:
    n = collatz(int(n))
````

### Nested Loops

````python
# Take user input
number = 3

# condition of the while loop
while number < 5 :  # defines the range
    # condition of the nested while loop    
    while number % 2 == 0: 
        print("The number "+ str(number)+" is even")
        break
        
    number +=1
````

## Pandas replacing using Regex

````python
olympic_df = pd.read_csv('olympics_cleaned.csv')
````

````python
olympic_df.tail(5)
````

````python
# take copy of the dataframe
import copy

olympic_df_strip_codes = copy.deepcopy(olympic_df)
````

**regex used**

* ```[a-zA-Z]``` - any letters a-z upper or lower
* ```\[``` - square bracket- escaped
* ```\]``` - square bracket - escaped
* ```\)``` - bracket - escaped
* ```\(``` - bracket - escaped

````python
olympic_df_strip_codes.tail(5)
````

````python
# remove extra country codes in square brackets '[]

olympic_df_strip_codes.countries = olympic_df_strip_codes.countries.str.replace(r'\([a-zA-Z]+\)\s\[[a-zA-Z]+\]', '', regex=True)
olympic_df_strip_codes.tail(5)
````

````python
# remove extra country codes in  brackets

olympic_df_strip_codes.countries= olympic_df_strip_codes.countries.str.replace(r'\([a-zA-Z]+\)', '', regex=True)
olympic_df_strip_codes.tail(5)
````

**Drop Total row**

````python
# drop the last row

olympic_df_strip_codes.drop(146, axis=0).tail() #row label, axis=1
````

## Pandas splitting

````python
olympic_df.tail(5)
````

````python
# take copy of the dataframe
import copy
olympic_df_split_codes = copy.deepcopy(olympic_df)
````

````python
olympic_df_split_codes[['countries','codes']] = olympic_df_split_codes['countries'].str.split('(', expand=True)
````

````python
olympic_df_split_codes.tail()
````

**Let's split the extra codes again**

````python
# take copy of the dataframe

olympic_df_split_codes_extra = copy.deepcopy(olympic_df_split_codes)
````

````python
# split the extra codes again

olympic_df_split_codes_extra[['codes','ext_codes']] = olympic_df_split_codes_extra['codes'].str.split(')', expand=True)
````

````python
olympic_df_split_codes_extra.tail()
````

## Rearranging columns

````python
olympic_df_col_arranged = copy.deepcopy(olympic_df_split_codes_extra)
````

````python
olympic_df_col_arranged.head(1)
````

````python
# store the codes list/column separate

codes = olympic_df_col_arranged.codes
ext_codes = olympic_df_col_arranged.ext_codes
````

````python
# drop the two codes columns

olympic_df_col_arranged.drop(columns=['codes','ext_codes'],inplace=True)
````

````python
olympic_df_col_arranged.head()
````
